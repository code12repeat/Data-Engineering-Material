```python
from pyhive import hive

#define connection parameter
host='localhost' #Since hive is running localy
port=10000 # Default port for HiveServer2
username='maneelcha49dgre'
database='default' #use Default or give your database name

#Establieh a connection
conn= hive.Connection(host=host,port=port,username=username,database=database)

#create a cursor object
cursor=conn.cursor()
cursor.execute("SHOW TABLES")

results=cursor.fetchall()

#print result
for row in results:
    print(row)
    
#close the cursor and connection
cursor.close()
conn.close()
```

```python
from sqlalchemy.engine import create_engine
import pandas as pd

host='localhost' #Since hive is running localy
port=10000 # Default port for HiveServer2
username='maneelcha49dgre'
database='default'

connection_string=f'hive://{username}@{host}:{port}/{database}'

engine=create_engine(connection_string)

query="SHOW TABLES"

df=pd.read_sql(query,engine)

print(df)
```
