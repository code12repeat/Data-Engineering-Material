#### **Set**

What is a Set?
- A set in Python is an unordered collection of unique elements.
- Sets are defined by enclosing elements in curly braces {}, separated by commas
- Sets automatically remove duplicate values, ensuring all elements are unique.



````python
# You can create a set by listing elements inside curly braces {}.
# Example of creating a set
my_set = {"P1001", "P1002", "P1001", "P1003", "P1004"}
````

**Adding Elements**: Use the add() method to add a single element to a set.



````python
my_set.add("P1005")
print(my_set)
````

**Removing Elements**: Use the remove() or discard() method to remove an element from a set. discard() does not raise an error if the element is not present.

````python
my_set.remove("P1002")
print(my_set)
````

````python
my_set.discard("P1004")
print(my_set) # no error
````

**Set Operations**

- Union: Combines elements from two sets, excluding duplicates.
- Intersection: Returns elements common to both sets.
- Difference: Returns elements in one set but not the other.



````python
set1 = {"P1001", "P1002", "P1003"}
set2 = {"P1003", "P1004", "P1005"}
union_set = set1 | set2  # Using the | operator
intersection_set = set1 & set2  # Using the & operator
difference_set = set1 - set2  # Elements in set1 but not in set2

print("union_set", union_set)
print("intersection_set", intersection_set)
print("difference_set", difference_set)
````

### Conditional Statements



>Conditional statements allow you to execute different blocks of code based on specified conditions.

> Conditional statements are an important part of decision-making in Python. Results are generated for the satisfied condition.

**If Else Conditional Statement**

> The if-else statement in Python is used for decision-making. The general syntax of an if-else statement is:
```
    if condition:
         Block of code to execute if condition is True
    else:
         Block of code to execute if condition is False
```
>It contains a body of code which runs only when the condition given in the if statement is true. If the condition is false, then the optional else statement runs which contains some code for the else condition.


**if-elif-else Statement**

- The elif statement allows you to check multiple conditions sequentially.
- It executes the block associated with the first True condition encountered. If no condition is True and an else block is provided, that block is executed.

```
    if condition1:
        Block of code to execute if condition1 is True
    elif condition2:
        Block of code to execute if condition1 is False and condition2 is True
    else:
        Block of code to execute if all conditions are False
```

**Let's store the total sales for each product in a collection**



````python
# Initialize an empty dictionary to store total sales per product
total_sales_per_product = {}

# Iterate over each record in the sales data
for record in Sales_data:
    product_id = record["Product_ID"]
    sales_amount = record["Price"] * record["Quantity_Sold"]

    # If the product ID is already in the dictionary, add to its sales total
    if product_id in total_sales_per_product:
        total_sales_per_product[product_id] += sales_amount
    else:
        # Otherwise, initialize the total sales for this product ID
        total_sales_per_product[product_id] = sales_amount

# Display the total sales for each product
for product_id, total_sales in total_sales_per_product.items():
    print(f"Product ID {product_id} has total sales of: {total_sales:.2f} Rs")
````

**We can use similar approach to find out the total sales value for each category.**

````python
# Calculate total sales for each category
category_sales = {}

# Iterate over each record in Sales_data
for record in Sales_data:
    category = record["Category"]
    sales = record["Price"] * record["Quantity_Sold"]
    # Update category_sales dictionary with total sales for each category
    #If category exists in category_sales, it returns its current value.
    #If not (category is not yet in category_sales), it returns 0 (the default value specified as the second argument).
    category_sales[category] = category_sales.get(category, 0) + sales

# Print results
for category, total_sales in category_sales.items():
    print(f"Category: {category}, Total Sales: {total_sales:.2f} Rs")
````

````python

trending_products = {}
consistent_increase = set()

# Code to also track product names
product_names = {}

for record in Sales_data:
    product_id = record["Product_ID"]
    product_name = record["Product_Name"]
    quantity_sold = record["Quantity_Sold"]

    # Keep track of product names
    product_names[product_id] = product_name

    # Check if product_id is already in trending_products
    if product_id not in trending_products:
        trending_products[product_id] = [quantity_sold]
    else:
        # Check if current quantity_sold is greater than the last recorded quantity
        if quantity_sold > trending_products[product_id][-1]:
            consistent_increase.add(product_id)
        trending_products[product_id].append(quantity_sold)

# Filter products and print the results
for product_id, sales_history in trending_products.items():
    if len(sales_history) > 1 and product_id in consistent_increase:
        print(f"Product name {product_names[product_id]} shows a general increasing trend in sales quantity.")

if not consistent_increase:
    print("No products found with a consistent increasing sales quantity trend.")
````

* Now let's Extend the Product class by adding a method apply_discount which takes a percentage discount as an argument and adjusts the product's price accordingly. After applying the discount, the method should print the new price of the product.

````python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def __str__(self):
        return f"Product Name: {self.name}, Price: {self.price:.2f} Rs, Category: {self.category})"

    def apply_discount(self, discount_percentage):
        self.price *= (1 - discount_percentage / 100)
        print(f"New price of {self.name} after {discount_percentage}% discount: {self.price:.2f} Rs")

# Method (apply_discount):
#apply_discount(self, discount_percentage): Method to apply a discount to the product's price.
# Modifies self.price by reducing it based on the given discount_percentage.

# Example Usage
product = Product("Bluetooth Headphones", 1500.0, "Electronics")
print(product)
product.apply_discount(10)  # Applying a 10% discount
````

We can also try to create a SalesRecord class that manages the sales of various products. It should have methods to add a product sale and to calculate total sales. Each sale record should include the product, quantity sold, and calculate the total sale amount (price * quantity).

````python
class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def __str__(self):
        return f"Product Name: {self.name}, Price: {self.price:.2f} Rs, Category: {self.category}"

class SalesRecord:
    def __init__(self):
        self.sales = []  # Initialize an empty list to store sales records

    def add_sale(self, product, quantity):
        total_amount = product.price * quantity  # Calculate the total amount for this sale
        self.sales.append({"product": product, "quantity": quantity, "total_amount": total_amount})
        # Append a dictionary representing the sale to the `sales` list

    def calculate_total_sales(self):
        total_sales = sum(sale["total_amount"] for sale in self.sales)
        return total_sales  # Return the total sales amount

# Example Usage
sales_record = SalesRecord()  # Create an instance of the SalesRecord class

# Create instances of the Product class
product1 = Product("Organic Apple", 250, "Groceries")
product2 = Product("Yoga Mat", 400, "Sporting Goods")

# Add sales records to the sales_record instance
sales_record.add_sale(product1, 30)
sales_record.add_sale(product2, 10)

# Calculate and print the total sales amount recorded in the sales_record instance
total_sales_amount = sales_record.calculate_total_sales()
print(f"Total sales amount: {total_sales_amount:.2f} Rs")
````

**Q14. Design a simple Product class with attributes like name, price, category, and stock (representing the number of items in stock). Extend the Product class by adding a method update_stock which takes a number as an argument and updates the product's stock by adding this number (which could be positive or negative). The method should then print the updated stock level. If the stock level goes below zero, the method should print a warning message indicating a stockout.**

````python
class Product:
    def __init__(self, name, price, category, stock):
        self.name = name
        self.price = price
        self.category = category
        self.stock = stock

    def __str__(self):
        return f"Product(Name: {self.name}, Price: {self.price} Rs, Category: {self.category}, Stock: {self.stock})"

    def update_stock(self, quantity):
        self.stock += quantity
        if self.stock < 0:
            print(f"Warning: Stockout for {self.name}. Current stock is {self.stock}.")
        else:
            print(f"Updated stock for {self.name} is now {self.stock}.")

# Example Usage
product = Product("Organic Apple", 200, "Groceries", 500)
print(product)
product.update_stock(-20)  # Selling 20 apples
product.update_stock(-35)  # Selling more than the current stock
````
