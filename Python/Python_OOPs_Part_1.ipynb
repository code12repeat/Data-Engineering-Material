**1. Introduction to Object Oriented Programming (OOPs)**

In the realm of programming, **procedural programming and object-oriented programming are the two paradigms** that are frequently employed in the creation of programmes.

**What is OOPs in Python?**
- Object-oriented programming (OOP) places a strong emphasis on using objects and classes to represent real-world entities and their connections.
- OOP in Python entails the definition of classes, which serve as templates for building objects with standardised traits and behaviours.
- Since objects are instances of classes, they are capable of having their own special methods and data attributes.
- Encapsulation, inheritance, and polymorphism, which enable code reuse, modularity, and flexibility in the design and maintenance of software systems, are the three main OOP ideas in Python.

**3. Adding parameters to using in-built classmethod()**

classmethod() is a built-in python function where the class method is returned for a specified function.

**Syntax:**
classmethod(function)

Parameter: The function name is accepted as parameter

Return Type:Coverted class method is returned as function


**Let's see a simple classmethod program**

```python
class college:
    course = 'BE'
    def opted(obj):
        print("Chosen course : ", obj.course)
college.opted = classmethod(college.opted)
college.opted()
```

In the above example,we created a class with name as "college" and member variable "course" then created a function "opted" which prints the object.

Then we passed the method "college.opted" into classmethod which converts the methods to a class method and then we call the class function "opted" without creating a function object.

**Let's see program to create class method using classmethod()**

```python
class employee:
	#variable
	empname = "Ram"
	#function
	def print_empname(obj):
		print("The employee name is : ", obj.empname)
# create print_empname classmethod
# before creating this line print_empname()
# It can be called only with object not with class
employee.print_empname = classmethod(employee.print_empname)
# now this method can be called as classmethod
# print_empname() method is called a class method
employee.print_empname()
```

**4. Constructors and Destructors in Python**


**Constructors:**
- Object instantiation will happen with the help of constructors.

- The work of constructors is to assign values to the data members of the class when an object of the class is created.

- init method is called constructor in Python,everytime object is created this method is called.

**Syntax**

Constructor.PNG

**Types of Constructors**
- **Default Constructors:**
These constructors does not accept any arguments.One reference to the instance being formed serves as the sole argument its definition.

   **Let's checkout a example demonstrating default constructors**





```python
class employee:
	# default constructor
	def __init__(self):
		self.empname = "Tarun"
	# a method for printing data members
	def print_empname(self):
		print(self.empname)
# creating object of the class
obj = employee()
# calling the instance method using the object obj
obj.print_empname()
```

- **Parameterized Constructors:**
These constructors contains parameters,it requires a reference to the instance being created, known as self, as its first argument, and the programmer supplies the remaining arguments.

   **Let's checkout an example where we multiply two values and obtain their product using parameterized constructor**

```python
class Product:
	number1 = 0
	number2 = 0
	product = 0
	# parameterized constructor
	def __init__(self, n1, n2):
		self.number1 = n1
		self.number2 = n2
	def display(self):
		print("Enter First number = " + str(self.number1))
		print("Enter Second number = " + str(self.number2))
		print("Product of two numbers = " + str(self.product))
	def calc(self):
		self.product = self.number1 * self.number2
# creating object of the class
# this will invoke parameterized constructor
obj1 = Product(1000, 2000)
# creating second object of same class
obj2 = Product(10, 20)
# perform Multiplication on obj1
obj1.calc()
# perform Multiplication  on obj2
obj2.calc()
# display result of obj1
obj1.display()
# display result of obj2
obj2.display()
```

**Advantages of using constructors in Python:**
- Constructors are easy to implement in Python.
- They are used to initialize the Objects of the class which makes the code visibility clear.  
- They are also used to enforce encapsulation



**Disadvantages of using constructors in Python:**

- A single class cannot contain mutiple constructors with different parameters, which results in no Overriding.  
- They are having limited functionality, for example access modifiers are absent in constructors.

**Destructors:**

When an object is destroyed, the destructors are called. Destructors are not as necessary in Python as they are in C++ since Python contains a garbage collector that manages memory automatically.

**Notice that when an object is no longer referenced or when the programme terminates, a reference to that object is likewise destroyed.**

**Syntax**

Destructors.PNG

**Let's see a simple example of destructor.**



```python
class Student:
	# Initializing
	def __init__(self):
		print('Student created.')
	# Deleting (Calling destructor)
	def __del__(self):
		print('Destructor was called hence Student deleted.')
obj = Student()
del obj
```

In the above program the destructor was automatically called since we destroyed all references to the object "obj" with the del keyword.

**Advantages of Destructors**
- Destructors offer automatic resource cleanup when an object is no longer required.
- No matter how or when a object is utilised, destroyers make sure that it is properly cleaned up.
- Destructors are used for debugging as they are involved in tracking of object's lifespan and ascertain when it is being destroyed.


**Disadvantages of Destructors**
- The exact timing of when a destructor is called is not predictable in Python.
- Defining a destructor in Python can have a performance overhead, especially if the destructor performs complex or time-consuming cleanup operations.
- Debugging code that uses destructors can be difficult because it is not always clear when a destructor will be called.

**5. Class constructor in Python**

In Python, a class constructor is a special method that is called when an object of the class is created.

The constructor method is named __init__() and it is used to initialize the attributes of the object.

```python
# Class with init method
class student:
	# using init method
	def __init__(self, stuname):
		self.stuname = stuname
	# Sample method
	def say_hello(self):
		print('Hello, my name is', self.stuname)
p = student('Shiv')
p.say_hello()
```

- The self parameter is a reference to the object being created.

- Inside the constructor, we set the name as attribute of the object using the value passed in as parameter.

- The say_hello() method is a simple method that prints out a greeting message using the name attribute.

**6. Class and Interface**

In Python, classes and interfaces are used to define custom data types and organize related functionality.




**Class:**
- Class serve as model for building items.
- They offer a means of combining data and processes.
- The characteristics and methods that objects will have are described in the class definition.
- Whereas methods are functions that can change the data or carry out other operations, attributes are variables that store data.

**Here is a simple example**

```python
# define a class
class Bike:
    bikename = ""
    bikegear = 0
# create object of class
bike1 = Bike()
# access attributes and assign new values
bike1.bikegear = 5
bike1.bikename = "Royal Enfield"
print(f"Name: {bike1.bikename}, Gears: {bike1.bikegear} ")
```

**Interface:**
- Interfaces are a bit different from classes.
- They are used to define a set of methods that a class should implement.
- They don't provide any implementation details themselves, they simply define the contract that a class should follow.
- In Python, interfaces are usually defined using abstract base classes (ABCs) from the abc module.

**Here is a simple example demonstrating one interface and two sub classes to that interface**

```python
from abc import *
class Database(ABC):
   @abstractmethod
   def connect(self):
       pass
   @abstractmethod
   def disconnect(self):
       pass

class MySQL(Database):
   def connect(self):
       print('Connecting to MySQL Database...')
   def disconnect(self):
       print('Disconnecting to MySQL Database...')

class MS_SQLServer(Database):
   def connect(self):
       print('Connecting to MS_SQL Server Database...')
   def disconnect(self):
       print('Disconnecting to MS_SQL Server Database...')

dbname=input('Enter Database Name either MySQL or MS_SQLServer:')
classname=globals()[dbname]
x=classname()
x.connect()
x.disconnect()
```
