**3. Adding parameters to using in-built classmethod()**
* A classmethod is a method that is bound to the class and not the instance of the class. It can access and modify class state that applies across all instances of the class.

* Class methods are used to access or modify the class state that applies to all instances.

* To define a class method, you use the @classmethod decorator or the classmethod() function.

* classmethod() is a built-in python function where the class method is returned for a specified function.

**Syntax:**
classmethod(function)

Parameter: The function name is accepted as parameter

Return Type:Coverted class method is returned as function

**Class Methods vs Instance Methods:**

> Instance methods take a parameter called self which refers to the instance of the class.

> Class methods take a parameter called cls which refers to the class itself.


**Let's see a simple classmethod program**

```python
class college:
    # Class attribute shared by all instances
    course = 'BE'

    # Define a method that takes an instance as the first parameter
    def opted(obj):
        # Print the chosen course, accessed through the instance
        print("Chosen course : ", obj.course)

# Convert the opted method into a class method
college.opted = classmethod(college.opted)

# Call the class method without needing to instantiate the class
college.opted()
```

In the above example,we created a class with name as "college" and member variable "course" then created a function "opted" which prints the object.

Then we passed the method "college.opted" into classmethod which converts the methods to a class method and then we call the class function "opted" without creating a function object.

**Let's see program to create class method using classmethod()**

```python
class College:
    course = 'BE'  # Class attribute shared by all instances

    def __init__(self, name):
        self.name = name  # Instance attribute unique to each instance

    def show_details(self):
        print(f"Name: {self.name}, Course: {self.course}")

    # Define a method and later convert it to a class method
    def change_course(cls, new_course):
        cls.course = new_course  # Change the class attribute for all instances

    change_course = classmethod(change_course)  # Convert the method to a class method

# Create instances of the College class
student1 = College('Alice')
student2 = College('Bob')

# Display details before changing the course
student1.show_details()
student2.show_details()
# Change the course using the class method
College.change_course('BTech')

# Display details after changing the course
student1.show_details()
student2.show_details()
```

**Using @classmethod decorator**

> Class Attribute: course is a class attribute shared by all instances.

> Instance Attribute: name is an instance attribute unique to each instance.

> Class Method: change_course is a class method that modifies the class attribute course.
*italicized text*
> Instance Method: show_details is an instance method that prints instance and class attributes

```python
class College:
    course = 'BE'  # Class attribute shared by all instances

    def __init__(self, name):
        self.name = name  # Instance attribute unique to each instance

    @classmethod
    def change_course(cls, new_course):
        cls.course = new_course  # Change the class attribute for all instances

    def show_details(self):
        print(f"Name: {self.name}, Course: {self.course}")

# Create instances of the College class
student1 = College('Alice')
student2 = College('Bob')

# Display details before changing the course
student1.show_details()
student2.show_details()

# Change the course using the class method
College.change_course('BTech')

# Display details after changing the course
student1.show_details()
student2.show_details()
```

**Static method**

In object-oriented programming (OOP), static methods are methods that belong to a class rather than an instance of the class. Unlike instance methods, they do not operate on instances of the class (self) or have access to instance attributes. Instead, static methods are associated with the class itself and can be called without creating an instance of the class.

* Belong to the Class: Static methods are defined within a class but do not operate on instances (self).

* No Access to Instance Attributes: They cannot access or modify instance attributes or instance methods directly.

* Decorated with @staticmethod: In Python, static methods are explicitly marked using the @staticmethod decorator.

**Advantages of Static Methods:**
* Clearly signals that the method does not modify instance state, making code easier to understand.
* Useful for defining standalone functions related to the class but not dependent on instance-specific data.
* Since they do not require instance creation (ex : without creating instances (Calculator())), they can be slightly faster in execution.

```python
class Calculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        return self.x + self.y

    @staticmethod
    def add_static(a, b):
        return a + b

# Example usage
calc = Calculator(5, 3)
print(calc.add())             # Output: 8 (using instance method)
print(Calculator.add_static(10, 3))  # Output: 8 (using static method)
```

**Instance Method (add):**

> An instance method like add is defined within the class using def add(self):.
> It operates on instance-specific data accessed via self (e.g., self.x, self.y).

> calc = Calculator(5, 3) creates an instance calc with x = 5 and y = 3.
> calc.add() is called using the instance calc, which accesses self.x and self.y to perform addition.

> Output: 8 (sum of 5 and 3)

**Static Method (add_static):**

> Static methods are defined using @staticmethod decorator.
> They do not take self parameter and do not access or modify instance-specific data (self.x, self.y).
Usage:

> Calculator.add_static(10, 3) calls the static method directly on the class Calculator.

> It does not require an instance (calc) and operates solely on the parameters a and b.

> Output: 13 (sum of 10 and 3).


> Using both instance methods and static methods helps in maintaining clear and readable code.
Instance methods clearly show they operate on instance attributes (self.x, self.y), while static methods are standalone and do not depend on instances.


**4. Constructors and Destructors in Python**


**Constructors:**
- Object instantiation will happen with the help of constructors.

- The work of constructors is to assign values to the data members of the class when an object of the class is created.

- __init__ is the constructor method. It is called automatically when an instance of the class is created.

**Syntax**

Constructor.PNG

**Types of Constructors**
- **Default Constructors:**
These constructors does not accept any arguments.One reference to the instance being formed serves as the sole argument its definition.

   **Let's checkout a example demonstrating default constructors**





```python
class Student:
	# default constructor
	def __init__(self):
		self.name = "Tarun"

	# a method for printing data members
	def print_name(self):
		print(self.name)

# creating object of the class
obj = Student()

# calling the instance method using the object obj
obj.print_name()
```

- **Parameterized Constructors:**
These constructors contains parameters,it requires a reference to the instance being created, known as self, as its first argument, and the programmer supplies the remaining arguments.

  

```python
class Student:
    def __init__(self, name, age, grade):
        # The constructor method to initialize attributes
        self.name = name
        self.age = age
        self.grade = grade

    def display_info(self):
        # Method to display student information
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")

# Creating instances of the Student class
student1 = Student("Alice", 20, "A")
student2 = Student("Bob", 22, "B")

# Displaying information for both students
student1.display_info()
student2.display_info()
```

 **Let's checkout an example where we multiply two values and obtain their product using parameterized constructor**

```python
class Product:
    # Class attributes, shared by all instances
    number1 = 0
    number2 = 0
    product = 0

    # Parameterized constructor
    def __init__(self, n1, n2):
        """
        Initializes the Product instance with two numbers.

        Parameters:
        n1 (int or float): First number.
        n2 (int or float): Second number.
        """
        self.number1 = n1
        self.number2 = n2

    def display(self):
        """Displays the numbers and their product."""
        print("First number = " + str(self.number1))
        print("Second number = " + str(self.number2))
        print("Product of two numbers = " + str(self.product))

    def calc(self):
        """Calculates the product of the two numbers and stores it in the product attribute."""
        self.product = self.number1 * self.number2

# Creating an object of the Product class with numbers 1000 and 2000
obj1 = Product(1000, 2000)

# Creating another object of the Product class with numbers 10 and 20
obj2 = Product(10, 20)

# Perform multiplication on obj1 and obj2
obj1.calc()
obj2.calc()

# Display the result of obj1
obj1.display()

# Display the result of obj2
obj2.display()
```

**Advantages of using constructors in Python:**
- Constructors are easy to implement in Python.
- They are used to initialize the Objects of the class which makes the code visibility clear.  
- They are also used to enforce encapsulation



**Disadvantages of using constructors in Python:**

- A single class cannot contain mutiple constructors with different parameters, which results in no Overriding.  
- They are having limited functionality, for example access modifiers are absent in constructors.

**Destructors:**

When an object is destroyed, the destructors are called. Destructors are not as necessary in Python as they are in C++ since Python contains a garbage collector that manages memory automatically.

They are defined using the __del__() method.Used to perform cleanup tasks or release resources held by an object before it is destroyed.

**Notice that when an object is no longer referenced or when the programme terminates, a reference to that object is likewise destroyed.**

**Syntax**

Destructors.PNG

**Let's see a simple example of destructor.**



```python
class Student:
	# Initializing
	def __init__(self):
		print('Student created.')
	# Deleting (Calling destructor)
	def __del__(self):
		print('Destructor was called hence Student deleted.')
obj = Student()
del obj
```

In the above program the destructor was automatically called since we destroyed all references to the object "obj" with the del keyword.

**Advantages of Destructors**
- Destructors offer automatic resource cleanup when an object is no longer required.
- No matter how or when a object is utilised, destroyers make sure that it is properly cleaned up.
- Destructors are used for debugging as they are involved in tracking of object's lifespan and ascertain when it is being destroyed.


**Disadvantages of Destructors**
- The exact timing of when a destructor is called is not predictable in Python.
- Defining a destructor in Python can have a performance overhead, especially if the destructor performs complex or time-consuming cleanup operations.
- Debugging code that uses destructors can be difficult because it is not always clear when a destructor will be called.

**6. Class and Interface**

In Python, classes and interfaces are used to define custom data types and organize related functionality.




**Class:**

**Here is a simple example**

```python
# define a class
class Bike:
    bikename = ""
    bikegear = 0
# create object of class
bike1 = Bike()
# access attributes and assign new values
bike1.bikegear = 5
bike1.bikename = "Royal Enfield"
print(f"Name: {bike1.bikename}, Gears: {bike1.bikegear} ")
```

**Interface:**
- Interfaces are a bit different from classes.
- They are used to define a set of methods that a class should implement.
- They don't provide any implementation details themselves, they simply define the contract that a class should follow.
- In Python, interfaces are usually defined using abstract base classes (ABCs) from the abc module.

**Here is a simple example demonstrating one interface and two sub classes to that interface**

```python
from abc import *
class Database(ABC):
   @abstractmethod
   def connect(self):
       pass
   @abstractmethod
   def disconnect(self):
       pass

class MySQL(Database):
   def connect(self):
       print('Connecting to MySQL Database...')
   def disconnect(self):
       print('Disconnecting to MySQL Database...')

class MS_SQLServer(Database):
   def connect(self):
       print('Connecting to MS_SQL Server Database...')
   def disconnect(self):
       print('Disconnecting to MS_SQL Server Database...')

dbname=input('Enter Database Name either MySQL or MS_SQLServer:')
classname=globals()[dbname]
x=classname()
x.connect()
x.disconnect()
```

```python
# Abstract base class defining a Shape interface
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

# Concrete subclass implementing the Shape interface
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius

# Usage of Circle class
circle = Circle(5)
print("Area of Circle:", circle.area())
print("Perimeter of Circle:", circle.perimeter())
```
