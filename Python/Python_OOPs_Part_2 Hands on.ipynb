
### **3. Changing Class Members in Python**

Changing class members refers to modifying attributes or methods of a class after it has been defined. In Python, you can change class members dynamically, which can be useful in certain situations such as updating default values, adding new methods, or altering existing ones.

Python allows us to change class members in a variety of ways.

**Here are a few typical methods:**



**Changing class attributes:**

- We can directly change class attributes by giving them new values.

- Every instance of the class share the same class characteristics.

**For Example**


```python
class MyClass:
    count = 0   # class attribute
MyClass.count = 10   # modify class attribute

MyClass.count
```

**In the above example, the count class attribute is modified by assigning a new value to it.**

```python
class Car:
    color = "red"

# Modify class attribute
Car.color = "blue"

# Access modified attribute
print(Car.color)  # Output: blue
```

You can add new class attributes dynamically by assigning values to them using the class name.



```python
class Car:
    color = "red"

# Add a new class attribute
Car.brand = "Toyota"

# Access the new attribute
print(Car.brand)  # Output: Toyota
```

You can also modify methods by assigning a new function or method to the class attribute.



```python
class Car:
    def drive(self):
        print("Driving the car")

# Define a new method which is outside a class
def new_drive_method(self):
    print("Driving carefully")

# Modify the method of the class dynamically
Car.drive = new_drive_method

# Create an instance and call the modified method
my_car = Car()
my_car.drive()  # Output: Driving carefully
```

**Modifying instance attributes:**

We can modify instance attributes by accessing them through instances of the class. Instance attributes are unique to each instance of the class.

**For example:**


```python
class MyClass:
    def __init__(self):
        self.value = 0   # instance attribute
obj = MyClass()
obj.value = 10   # modify instance attribute
```

**In this example, the value instance attribute is modified by accessing it through an instance of the MyClass class.**


**Modifying class methods:**

We can modify class methods by redefining them in a subclass.

**For example:**

```python
class MyClass:
    def my_method(self):
        print("Hello, Everyone!")

class MySubclass(MyClass):
    def my_method(self):
        print("Thankyou!")
obj = MySubclass()
obj.my_method()
```

**In this example, the my_method class method is redefined in the MySubclass subclass, which modifies its behavior.**


**Modifying instance methods:**

We can modify instance methods by defining new instance methods and replacing the existing ones.

**For example:**

```python
class MyClass:
    def my_method(self):
        print("Hello, Everyone!")

obj = MyClass()
obj.my_method = lambda: print("Thankyou")
obj.my_method()
```

**In this example, the my_method instance method is modified by defining a new function using a lambda expression and assigning it to the my_method attribute of an instance of the MyClass class.**

### **Inheritance**
- Inheritance allows a new class (the derived or child class) to be based on an existing class (the base or parent class).

- In Python, inheritance is implemented using the class keyword and the name of the parent class in parentheses after the name of the child class.

- The child class inherits all the attributes and methods of the parent class and can also add its own attributes and methods.

- This makes code reuse and organization easier, as common functionality can be implemented in the parent class and specialized functionality can be added in the child class.

**Here's an example of inheritance in Python:**


**Using** Parent.__init__()**



```python
class Parent:
    def __init__(self, fname, year):
        self.fname = fname  # Initialize fname attribute
        self.year = year    # Initialize year attribute

    def view(self):
        print(self.fname, self.year)  # Display fname and year attributes

class Child(Parent):
    def __init__(self, fname, year):
        # Call parent class constructor to initialize fname and year attributes
        Parent.__init__(self, fname, year)

        self.inventorname = "Guido van Rossum"  # Initialize additional attribute

    def view(self):
        # Override parent's view method to include additional information
        print(self.fname, "was developed by", self.inventorname, "in", self.year)

# Create an instance of Child class
ob = Child("Python Programming", '1980')

# Call the view method of Child class to display information
ob.view()
```

**Using super()**

```python
class Parent:
    def __init__(self, fname, year):
        self.fname = fname  # Initialize fname attribute in Parent class
        self.year = year    # Initialize year attribute in Parent class

    def view(self):
        print(self.fname, self.year)  # Display fname and year attributes

# Child class inherits from Parent class
class Child(Parent):
    def __init__(self, fname, year):
        super().__init__(fname, year)  # Call Parent's constructor to initialize fname and year
        self.inventorname = "Guido van Rossum"  # Additional attribute specific to Child

    def view(self):
        # Override Parent's view method to include additional information
        print(self.fname, "was developed by", self.inventorname, "in", self.year)

# Create an instance of Child class
ob = Child("Python Programming", '1980')
ob.view()  # Output: Python Programming was developed by Guido van Rossum in 1980
```

**Explanation of Inheritance:**

- Inherited Attributes (fname and year): When you create an instance of Child, such as ob = Child("Python Programming", '1980'), you're passing "Python Programming" as fname and '1980' as year.

- Usage in Child Class: The Child class doesn't redefine fname and year but inherits them from Parent, allowing it to access and use these attributes directly.

- Method Overriding: While the Child class overrides the view() method of Parent to add more specific information (inventorname), the attributes fname and year are still inherited and accessible as they were defined in the Parent class.

Example 2

```python
# Parent class
class Shape:
    def __init__(self, name):
        self.name = name  # Initialize name attribute with the provided name

    def area(self):
        pass  # Placeholder for area calculation, to be overridden in subclasses

# Child class inheriting from Shape
class Circle(Shape):
    def __init__(self, name, radius):
        super().__init__(name)  # Call parent's constructor to initialize name attribute
        self.radius = radius  # Initialize radius specific to Circle

    def area(self):
        return 3.14 * self.radius * self.radius  # Calculate and return area of Circle

# Another child class inheriting from Shape
class Rectangle(Shape):
    def __init__(self, name, length, breadth):
        super().__init__(name)  # Call parent's constructor to initialize name attribute
        self.length = length  # Initialize length specific to Rectangle
        self.breadth = breadth  # Initialize breadth specific to Rectangle

    def area(self):
        return self.length * self.breadth  # Calculate and return area of Rectangle

# Using the classes

# Create an instance of Circle
circle = Circle("Circle", 5)

# Create an instance of Rectangle
rectangle = Rectangle("Rectangle", 4, 6)

# Calculating and displaying areas

# Print area of Circle
print(f"Area of {circle.name} is: {circle.area():.2f}")

# Print area of Rectangle
print(f"Area of {rectangle.name} is: {rectangle.area()}")
```

* **Shape Class**: Defines a parent class Shape with an initializer (__init__) that sets the name attribute and an abstract method area() which is meant to be overridden in subclasses.

* **Circle Class**: Inherits from Shape. Uses super() to initialize the name attribute inherited from Shape. Implements area() method to calculate the area of a circle using its radius

* **Rectangle Class**: Also inherits from Shape. Uses super() to initialize the name attribute inherited from Shape. Implements area() method to calculate the area of a rectangle using its length and breadth

*  The Shape class defines common attributes and behaviors (name attribute and area() method) that are shared by both Circle and Rectangle classes. By inheriting from Shape, both subclasses (Circle and Rectangle) inherit these attributes and methods without needing to redefine them. This promotes code reuse and avoids redundancy.


**There are several types of inheritance based on how subclasses inherit from superclasses:**


1. **Single Inheritance:**

- In this type, a subclass inherits from only one superclass.

2. **Multiple Inheritance:**

- Here, a subclass inherits from more than one superclass.

3. **Multilevel Inheritance:**

In this type, a subclass inherits from a superclass, and another subclass inherits from the derived class.




**Single Inheritance**


```python
# Single Inheritance
class Parent:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Name:", self.name)

# Child inherits from Parent
class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def display_age(self):
        print("Age:", self.age)

# Creating objects for single inheritance
child = Child("Alice", 12)
child.display()
child.display_age()
```

> Child inherits from Parent.
Child extends Parent by adding an age attribute and display_age method.
Demonstrates super() usage to call the parent class constructor.


**Multiple Inheritance**

```python
# Multiple Inheritance with parameters
class A:
    def __init__(self, name):
        self.name = name

    def methodA(self):
        print("Method A from class A")

class B:
    def __init__(self, age):
        self.age = age

    def methodB(self):
        print("Method B from class B")

# C inherits from both A and B
class C(A, B):
    def __init__(self, name, age, salary):
        A.__init__(self, name)  # Initialize name using A's constructor
        B.__init__(self, age)   # Initialize age using B's constructor
        self.salary = salary    # Initialize salary locally

    def methodC(self):
        print("Method C from class C")

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Salary: {self.salary}")

# Creating objects for multiple inheritance with parameters
objC = C("Alice", 30, 50000)

objC.methodA()
objC.methodB()
objC.methodC()
objC.display()
```

Class A:
Defines methodA and initializes name in its constructor.

Class B:
Defines methodB and initializes age in its constructor.
Class C (Multiple Inheritance):

Inherits from both A and B.
Initializes name using A's constructor and age using B's constructor.


**Multilevel Inheritance**

```python
# Multilevel Inheritance with parameters
class A:
    def __init__(self, name):
        self.name = name

    def methodA(self):
        print("Method A from class A")

class B(A):
    def __init__(self, name, age):
        super().__init__(name)  # Initialize name using A's constructor
        self.age = age

    def methodB(self):
        print("Method B from class B")

class C(B):
    def __init__(self, name, age, salary):
        super().__init__(name, age)  # Initialize name and age using B's constructor
        self.salary = salary

    def methodC(self):
        print("Method C from class C")

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Salary: {self.salary}")

# Creating objects for multilevel inheritance with parameters
objC = C("Bob", 35, 60000)
objC.methodA()

objC.methodB()
objC.methodC()

objC.display()

```

Class A:
Defines methodA and initializes name in its constructor.

Class B (Inherits from A):
Inherits methodA and name attribute from A.
Initializes age in its constructor.

Class C (Inherits from B):Inherits methodA, methodB, name, and age from B.
Initializes salary in its constructor.


### **Encapsulation**

- Encapsulation in Python is a concept of object-oriented programming (OOP) that refers to **the process of hiding the implementation details of an object from the outside world**, while providing an interface to access the object's properties and methods.

- In Python, encapsulation is achieved using access modifiers such as public, private, and protected. However, unlike other languages such as Java, **Python does not have strict rules for access modifiers**.

- Instead, Python uses naming conventions to indicate the level of access to an object's attributes and methods.


**Naming conventions of access specifiers:**

**Private:**
- The naming convention for private attributes and methods in Python is to prefix the attribute or method name with a double underscore (__).

- This makes it harder to access the attribute or method from outside the object. However, it is still possible to access private attributes and methods using name **mangling**, which involves prefixing the attribute or method name with the class name and a single underscore (_).

**Protected**

- Protected attributes and methods in Python are indicated by prefixing the attribute or method name with a single underscore (_)

- This convention indicates that the attribute or method is intended to be used only within the class and its subclasses, but it can still be accessed from outside the class.

**Public**

- Public attributes and methods in Python do not require any special naming convention.
- They can be accessed from anywhere in the program.

```python
class Person:
    def __init__(self, name, age):
        self.name = name      # Public attribute
        self.__age = age      # Private attribute

    def get_age(self):
        return self.__age     # Method to access private attribute

    def set_age(self, age):
        # Method to update private attribute with validation
        if age > 0 and age < 120:
            self.__age = age

    def display(self):
        # Method to display the person's details
        print(f"Name: {self.name}, Age: {self.__age}")

# Creating an instance of the Person class
person = Person("John", 30)

# Display initial details
person.display() # first output is generated by the first call to the display method, reflecting the initial state of the person object.

# Update the age using the setter method
person.set_age(35)

# Display updated details
person.display() # second output is generated by the second call to the display method, showing the updated state of the person object after changing the age using the set_age method.


# Access the age using the getter method
print(person.get_age()) # Third output is generated by calling the get_age method, which returns the current value of the private __age attribute.
```

In this example, the age attribute is marked as private by prefixing its name with a double underscore. The **get_age()** and **set_age()** methods are used to access and modify the age attribute, respectively. The display() method is used to display the object's name and age attributes.

Note that even though the age attribute is marked as private, it can still be accessed and modified using the **get_age()** and **set_age()** methods. However, attempting to access the attribute directly using **person.__age** will result in an AttributeError.

Example with Subclass Utilizing Protected Member


```python
class Employee:
    def __init__(self, name, salary, age):
        self.name = name          # Public attribute
        self._salary = salary     # Protected attribute
        self.__age = age          # Private attribute

    # Getter method for the private attribute
    def get_age(self):
        return self.__age

    # Setter method for the private attribute with validation
    def set_age(self, age):
        if 18 <= age <= 65:
            self.__age = age
        else:
            print("Please enter a valid age (18-65).")

    # Method to display the current state of the object
    def display(self):
        print(f"Name: {self.name}, Salary: {self._salary}, Age: {self.__age}")

# Subclass of Employee
class Manager(Employee):
    def __init__(self, name, salary, age, department):
        super().__init__(name, salary, age)
        self.department = department  # Additional attribute for Manager

    # Method to display manager details including department
    def display_manager_details(self):
        self.display()  # Calling the parent class's display method
        print(f"Department: {self.department}")

    # Method to give a raise to the manager
    def give_raise(self, percentage):
        self._salary += self._salary * (percentage / 100)

        # self._salary accesses the protected _salary attribute inherited from the Employee class.
        print(f"{self.name}'s new salary after {percentage}% raise: {self._salary}")

# Creating an instance of Manager
manager = Manager("Bob", 80000, 40, "IT")

# Displaying initial details of the manager
manager.display_manager_details()


# Giving a raise to the manager
manager.give_raise(10)
# Output: Bob's new salary after 10% raise: 88000.0

# Displaying updated details of the manager
manager.display_manager_details()


# Accessing the age using the getter method from parent class
print(manager.get_age())  # Output: 40
```

**Employee Class**: Defines a public attribute name and a protected attribute _salary.

**Manager Class (Subclass of Employee)**: Inherits from Employee and adds a public attribute department

The _salary attribute is defined with a single underscore (_) prefix, indicating it's intended to be a protected member. It stores salary information, and by convention, it should only be accessed or modified within the Employee class or its subclasses (Manager).

**Manager Class (Subclass of Employee):**
- Inherits from Employee and utilizes the _salary attribute directly within its display() method through super():

**Although protected members are intended for internal use within the class and its subclasses, Python's dynamic nature allows access to _salary outside the class hierarchy as well, such as in:**

```python
mgr = Manager("Jane Smith", 80000, 70,"Engineering")
print(f"Manager's salary: {mgr._salary}")  # Output: Manager's salary: 80000
```

Despite being accessible outside the class hierarchy, using protected members directly outside their intended scope can lead to unintended side effects or violate encapsulation principles.

**Private member**

```python
class Employee:
    def __init__(self, name, salary, age):
        self.name = name          # Public attribute
        self._salary = salary     # Protected attribute
        self.__age = age          # Private attribute

    # Getter method for the private attribute
    def get_age(self):
        return self.__age

    # Setter method for the private attribute with validation
    def set_age(self, age):
        if 18 <= age <= 65:
            self.__age = age
        else:
            print("Please enter a valid age (18-65).")

    # Method to display the current state of the object
    def display(self):
        print(f"Name: {self.name}, Salary: {self._salary}, Age: {self.__age}")

# Subclass of Employee
class Manager(Employee):
    def __init__(self, name, salary, age, department):
        super().__init__(name, salary, age)
        self.department = department  # Additional attribute for Manager

    # Method to display manager details including department
    def display_manager_details(self):
        self.display()  # Calling the parent class's display method
        print(f"Department: {self.department}")

    # Method to try accessing the private __age attribute (which will fail)
    def try_access_private_age(self):
        try:
            # Attempting to access __age directly from subclass (this will raise an AttributeError)
            print(self.__age)  # This will cause an AttributeError
        except AttributeError:
            print("AttributeError: Private attribute __age cannot be accessed directly.")

        # Accessing __age using the getter method from parent class
        print(f"Age from getter method: {self.get_age()}")
```

```python
# Creating an instance of Manager
manager = Manager("Alice", 75000, 45, "HR")

# Displaying manager details
manager.display_manager_details()

# Attempting to access __age directly from outside the class (this will raise an AttributeError)
try:
    print(manager.__age)  # This will cause an AttributeError
except AttributeError:
    print("AttributeError: Private attribute __age cannot be accessed directly from outside the class.")

# Trying to access __age within the subclass method
manager.try_access_private_age()
#   AttributeError:

# Trying to set age using the setter method (with validation)
manager.set_age(50)
manager.display()

# Trying to set an invalid age (this will not change the age)
manager.set_age(70)
manager.display()
```

Accessing __age Outside the Class: Directly accessing manager.__age from outside the class or even within the Manager subclass (print(self.__age)) raises an AttributeError. This demonstrates that __age is truly encapsulated and not accessible externally.

### **Polymorphism**

- Polymorphism is taken from the Greek words Poly (many) and morphism (forms). It means that the same function name can be used for different types.

**Example in Everyday Terms:**
> Think of a common scenario where multiple vehicles (cars, bicycles, trucks) can all "move". While the exact method of movement (driving, pedaling, hauling) differs between these vehicles, they all share the common concept of "movement".

- Python supports polymorphism through several mechanisms:
   - Duck Typing
   - Operator Overloading
   - Method overriding

**Duck typing:**

- Duck typing is a specific form of polymorphism in Python. It emphasizes the object's behavior over its actual type or class. The term comes from the phrase "If it looks like a duck, swims like a duck, and quacks like a duck, then it probably is a duck."

- Duck typing is a technique in which the type of an object is determined by its behavior rather than its type.

- If an object supports the methods or attributes that are required by a function or method, it can be used in place of an object of a specific type.

**For example:**

```python
#Duck typing demonstration
class sample:
	def __len__(self):
		return 30
if __name__ == "__main__":
	str = sample()
	print(len(str))
```

In this instance, we call the function len(), which returns the result of the __len__ method. Here, the class sample's attribute is defined via the __len__ method.

- We do not declare the parameter in method prototypes since the type of the object itself is not important in this.

- This implies that type checking cannot be done by compilers.

- So, whether an object has specific qualities at run time is what counts most. Dynamic languages thus implement duck typing.

- However, several static languages, like Haskell, now support it as well. But, Java/C# do not yet have this capability.


Example for Duck typing

```python
class Duck:
    def sound(self):
        print("Quack quack!")

class Dog:
    def sound(self):
        print("Woof woof!")

# Function that demonstrates duck typing
def make_sound(animal):
    animal.sound()

# Using duck typing with different objects
duck = Duck()
dog = Dog()

make_sound(duck)
make_sound(dog)
```

> In the above example Both Duck and Dog classes define a method sound().
They are independent of each other and do not share a common base class explicitly in this example.

> Function make_sound():
This function accepts a parameter animal.
It assumes animal has a sound() method and calls it without checking its specific type.

> Duck Typing:

> make_sound(duck): Here, duck is an instance of Duck, and make_sound() successfully calls duck.sound(), printing "Quack quack!".

> make_sound(dog): Similarly, dog is an instance of Dog, and make_sound() calls dog.sound(), printing "Woof woof!".

Duck typing focuses on the presence of specific methods (sound() in this case) rather than the specific class (like Duck or Dog) or type of the object.



**Operator overloading:**

- Operator overloading is a technique in which the behavior of an operator is defined for a class.

- When the operator is used with objects of the class, the overloaded behavior is executed instead of the default behavior.

- Operator overloading in Python allows operators such as +, -, *, /, ==, !=, etc., to behave differently based on the objects they are operating on. This means that the behavior of an operator can be customized for user-defined classes, enabling more intuitive and readable code.


```python
# # > Commonly Overloaded Operators:

# * Addition +  (__add__)
# * Subtraction - (__sub__)
# * Multiplication * (__mul__)
# * Division / (__truediv__ or __div__ in Python 2)
# * Equality == (__eq__)
# * Inequality != (__ne__)
# * Less than < (__lt__)
# * Greater than > (__gt__)  and many more...
```

**For example:**

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3.x, v3.y)
```

In this example, the Vector class overloads the addition operator (+) by defining the __add__() method. When the operator is used with instances of the Vector class, the overloaded behavior is executed, which adds the corresponding x and y values of the two vectors.

> Vector is a class that represents a mathematical vector with components x and y.

> Constructor (__init__ method):Initializes a Vector object with its components x and y.

> Operator Overloading (__add__ method):The __add__ method is a special method in Python that gets called when the + operator is used between two objects.
self refers to the instance of the Vector class on which the method is called (the left-hand side of the + operator).

> other refers to the object on the right-hand side of the + operator.
> Inside __add__, it creates and returns a new Vector object whose x component is the sum of self.x and other.x, and y component is the sum of self.y and other.y.

> When you use the + operator between two Vector objects (vector1 + vector2), Python automatically calls the __add__ method of the vector1 object, passing vector2 as other.

> vector1 + vector2 invokes vector1.__add__(vector2).

> Inside __add__, it computes x = 1 + 3 and y = 2 + 4, resulting in a new Vector object (4, 6).

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the addition operator (+)
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # Overloading the multiplication operator (*)
    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    # Overloading the equality operator (==)
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y

    # String representation of the Vector object
    def __str__(self):
        return f"({self.x}, {self.y})"

# Example usage of Vector class with operator overloading
v1 = Vector(2, 3)
v2 = Vector(5, 7)

# Overloaded addition operator (+)
v3 = v1 + v2
print("v1 + v2 =", v3)

# Overloaded multiplication operator (*)
v4 = v1 * 2
print("v1 * 2 =", v4)

# Overloaded equality operator (==)
print("v1 == v2:", v1 == v2)
```

**Method overriding**

Method overriding is a fundamental concept in object-oriented programming (OOP) where a subclass provides a specific implementation of a method that is already defined in its superclass. This allows a subclass to provide a specialized version of a method that is inherited from its superclass.

> In OOP, classes can inherit attributes and methods from other classes.
> A subclass (or child class) inherits from a superclass (or parent class).
Inherited Methods:

> When a subclass inherits a method from a superclass, it can choose to override (replace) that method with its own implementation.
Purpose:

> Method overriding allows you to provide specific behavior for a method in a subclass that is appropriate for that subclass, while still maintaining a common interface as defined in the superclass.



```python
# Superclass
class Animal:
    def sound(self):
        print("Generic animal sound")

# Subclass inheriting from Animal
class Dog(Animal):
    def sound(self):
        print("Bark bark!")

# Subclass inheriting from Animal
class Cat(Animal):
    def sound(self):
        print("Meow meow!")

# Creating instances and invoking overridden methods
animal = Animal()
dog = Dog()
cat = Cat()

animal.sound()
dog.sound()
cat.sound()
```

**In the above example**
* Animal Class: Defines a superclass Animal with a method sound() that prints a generic animal sound.

* Dog Class: Inherits from Animal and overrides the sound() method to print "Bark bark!" when called.

* Cat Class: Also inherits from Animal and overrides the sound() method to print "Meow meow!" when called.

* Method Invocation: When sound() is called on dog and cat instances, Python looks for the method in the respective subclass. If it finds an overridden method (sound() in Dog or Cat), it executes that specific implementation.

* More flexible, but less explicit about the structure.

**Encapsulation**

- Encapsulation in Python is a concept of object-oriented programming (OOP) that refers to **the process of hiding the implementation details of an object from the outside world**, while providing an interface to access the object's properties and methods.

- In Python, encapsulation is achieved using access modifiers such as public, private, and protected. However, unlike other languages such as Java, **Python does not have strict rules for access modifiers**.

- Instead, Python uses naming conventions to indicate the level of access to an object's attributes and methods.


**Naming conventions of access specifiers:**

**Private:**
- The naming convention for private attributes and methods in Python is to prefix the attribute or method name with a double underscore (__).

- This makes it harder to access the attribute or method from outside the object. However, it is still possible to access private attributes and methods using name **mangling**, which involves prefixing the attribute or method name with the class name and a single underscore (_).

**Protected**

- Protected attributes and methods in Python are indicated by prefixing the attribute or method name with a single underscore (_)

- This convention indicates that the attribute or method is intended to be used only within the class and its subclasses, but it can still be accessed from outside the class.

**Public**

- Public attributes and methods in Python do not require any special naming convention.
- They can be accessed from anywhere in the program.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age   # private attribute

    def get_age(self):
        return self.__age  # private method

    def set_age(self, age):
        if age > 0 and age < 120:
            self.__age = age  # private method

    def display(self):
        print(f"Name: {self.name}, Age: {self.__age}")

person = Person("John", 30)

person.display()
person.set_age(35)
person.display()
print(person.get_age())
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name      # Public attribute
        self.__age = age      # Private attribute

    def get_age(self):
        return self.__age     # Method to access private attribute

    def set_age(self, age):
        # Method to update private attribute with validation
        if age > 0 and age < 120:
            self.__age = age

    def display(self):
        # Method to display the person's details
        print(f"Name: {self.name}, Age: {self.__age}")

# Creating an instance of the Person class
person = Person("John", 30)

# Display initial details
person.display() # first output is generated by the first call to the display method, reflecting the initial state of the person object.

# Update the age using the setter method
person.set_age(35)

# Display updated details
person.display() # second output is generated by the second call to the display method, showing the updated state of the person object after changing the age using the set_age method.


# Access the age using the getter method
print(person.get_age()) # Third output is generated by calling the get_age method, which returns the current value of the private __age attribute.
```

In this example, the age attribute is marked as private by prefixing its name with a double underscore. The **get_age()** and **set_age()** methods are used to access and modify the age attribute, respectively. The display() method is used to display the object's name and age attributes.

Note that even though the age attribute is marked as private, it can still be accessed and modified using the **get_age()** and **set_age()** methods. However, attempting to access the attribute directly using **person.__age** will result in an AttributeError.

```python
class Employee:
    def __init__(self, name, salary, age):
        self.name = name          # Public attribute
        self._salary = salary     # Protected attribute
        self.__age = age          # Private attribute

    # Getter method for the private attribute
    def get_age(self):
        return self.__age

    # Setter method for the private attribute with validation
    def set_age(self, age):
        if 18 <= age <= 65:
            self.__age = age
        else:
            print("Please enter a valid age (18-65).")

    # Method to display the current state of the object
    def display(self):
        print(f"Name: {self.name}, Salary: {self._salary}, Age: {self.__age}")

# Subclass of Employee
class Manager(Employee):
    def __init__(self, name, salary, age, department):
        super().__init__(name, salary, age)
        self.department = department  # Additional attribute for Manager

    # Method to display manager details including department
    def display_manager_details(self):
        self.display()  # Calling the parent class's display method
        print(f"Department: {self.department}")

    # Method to give a raise to the manager
    def give_raise(self, percentage):
        self._salary += self._salary * (percentage / 100)

        # self._salary accesses the protected _salary attribute inherited from the Employee class.
        print(f"{self.name}'s new salary after {percentage}% raise: {self._salary}")

# Creating an instance of Manager
manager = Manager("Bob", 80000, 40, "IT")

# Displaying initial details of the manager
manager.display_manager_details()


# Giving a raise to the manager
manager.give_raise(10)
# Output: Bob's new salary after 10% raise: 88000.0

# Displaying updated details of the manager
manager.display_manager_details()


# Accessing the age using the getter method from parent class
print(manager.get_age())  # Output: 40
```

**Manager Class (Subclass of Employee)**: Inherits from Employee and adds a public attribute department

The _salary attribute is defined with a single underscore (_) prefix, indicating it's intended to be a protected member. It stores salary information, and by convention, it should only be accessed or modified within the Employee class or its subclasses (Manager).

### **Data Abstraction**

- Data abstraction refers to the process of hiding the complexity of an object's implementation by exposing only the essential details to the user.

- In Python, data abstraction is typically implemented using **abstract classes and interfaces.**


```python
# abstraction in python
from abc import ABC,abstractmethod

# abstract class
class Subject(ABC):
    @abstractmethod
    def subject(self):
        pass

class History(Subject):
    # override superclass method
    def subject(self):
        print("Subject is History")

class Civics(Subject):
    # override superclass method
    def subject(self):
        print("Subject is Civics")

class Geography(Subject):
    # override superclass method
    def subject(self):
        print("Subject is Geography")

class Economics(Subject):
    # override superclass method
    def subject(self):
        print("Subject is Economics")
history=History()
history.subject()
civics=Civics()
civics.subject()
geography=Geography()
geography.subject()
economics=Economics()
economics.subject()
```

- In the above python program, we created an abstract class Subject which extends Abstract Base Class (ABC).

- We also defined an abstract method subject. We also created other classes like History,Civics,Geography and Economics which all extend an abstract class Subject thus all these classes are subclasses and the Subject is the superclass.

- We provided different implementations for the subject method of all the subclasses.

**Example 2**


```python
from abc import ABC, abstractmethod

class Shape(ABC):  # Abstract class
    @abstractmethod
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def calculate_area(self):
        return self.length * self.breadth

# Usage
circle = Circle(5)
print(f"Area of Circle: {circle.calculate_area()}")  # Output: 78.5

rectangle = Rectangle(4, 6)
print(f"Area of Rectangle: {rectangle.calculate_area()}")  # Output: 24
```

* The Shape class is defined as an abstract class using ABC from the abc module. The @abstractmethod decorator is applied to calculate_area() method, indicating that any subclass of Shape must provide its own implementation of calculate_area().

* Both Circle and Rectangle are concrete subclasses of Shape and must implement calculate_area() method. This ensures that each shape calculates its area in a way appropriate to its shape type (Circle calculates based on radius, Rectangle based on length and breadth).

**Method overriding:**

- Method overriding is a technique in which a subclass provides a different implementation of a method that is already defined in its superclass.

- When the method is called on an object of the subclass, the implementation in the subclass is executed instead of the implementation in the superclass.

**For example:**

```python
from math import pi
class Shape:
    def __init__(self, name):
        self.name = name
    def area(self):
        pass
    def fact(self):
        return "I am a two-dimensional shape."
    def __str__(self):
        return self.name
class Square(Shape):
    def __init__(self, length):
        super().__init__("Square")
        self.length = length
    def area(self):
        return self.length**2
    def fact(self):
        return "Squares have each angle equal to 90 degrees."
class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius
    def area(self):
        return pi*self.radius**2
a = Square(2)
b = Circle(3)
print(b)
print(b.fact())
print(a.fact())
print(b.area())
```


- The fact() method for object a (of the Square class) is overridden due to polymorphism, which the Python interpreter immediately detects. It thus makes use of the one specified in the child class.


- The fact() function for object b, on the other hand, is called from the Parent Shape class because it isn't overridden.

**Method Overloading**

- Method overloading is a concept in object-oriented programming where two or more methods in a class have the same name but different parameters.

- In some programming languages, such as Java and C++, method overloading is supported natively.

- However, in **Python, method overloading is not supported** natively, but it can be achieved using some workarounds.

- In Python, a method is uniquely identified by its name, so it is not possible to have two methods with the same name and different parameters in the same class.

- However, there are a few ways to simulate method overloading in Python, such as using default arguments or variable-length argument lists.

**Here's an example of method overloading in Python using default arguments:**

```python
class MathOperations:
    def add(self, a, b, c=0):
        return a + b + c

math_ops = MathOperations()

print(math_ops.add(1, 2))       # Output: 3
print(math_ops.add(1, 2, 3))    # Output: 6
```

- In this example, the MathOperations class has a method called add() that takes two or three arguments.

- The third argument c has a default value of 0. This means that if the method is called with only two arguments, the third argument will be 0 by default.

- If the method is called with three arguments, the third argument will be the value provided by the caller.

**Another way to simulate method overloading in Python is to use variable-length argument lists.**

**Here's an example:**

```python
class MathOperations:
    def add(self, *args):
        result = 0
        for arg in args:
            result += arg
        return result

math_ops = MathOperations()

print(math_ops.add(1, 2))          # Output: 3
print(math_ops.add(1, 2, 3))       # Output: 6
print(math_ops.add(1, 2, 3, 4, 5)) # Output: 15
```

- In this example, the add() method takes a variable-length argument list using the *args syntax. The method then iterates over the arguments and adds them up to produce a result.
