## Distributions

**Kurtosis**

Skewness is a statistical measure that describes the degree of asymmetry in the distribution of data values around the mean.

A positive skew indicates a distribution with a longer tail on the right side, while a negative skew means a longer tail on the left.

Skewness is particularly useful in identifying potential outliers and understanding the distribution's shape, which can have implications for statistical analyses and modeling.

```python
mrp_skewness = df['Product_MRP'].skew()
mrp_skewness
```

**Positive Skewness**: If the result (value stored in skewness) is a positive number, it indicates that the distribution of 'Product_MRP' is skewed to the right. This means there are a relatively small number of very high prices pulling the mean higher than the median.

**Negative Skewness**: If the result is negative, the distribution is skewed to the left, meaning there are some very low prices that are pulling the mean lower than the median.

**Skewness Close to Zero**: If the skewness is close to zero, it suggests that the distribution of prices is fairly symmetrical around the mean.

**Kurtosis**

Kurtosis is a statistical measure used to describe the shape of a distribution's tails in relation to its peak.

A high kurtosis value indicates a distribution with heavy tails and a sharp peak (leptokurtic), suggesting the presence of outliers. Conversely, a low kurtosis (platykurtic) implies a distribution with light tails and a flatter peak, indicating a lack of outliers.

A kurtosis close to the value of a normal distribution (mesokurtic, typically around 3) suggests a distribution shape similar to the normal bell curve. This measure is particularly useful for identifying the extremity of outliers in data.

```python
mrp_kurtosis = df['Product_MRP'].kurtosis()
mrp_kurtosis
```

**High Kurtosis (>3)**: If the calculated kurtosis is significantly greater than 3, it indicates a leptokurtic distribution. This means the distribution has heavier tails and a sharper peak than the normal distribution. In the context of 'Product_MRP', this suggests more extreme values (either very high or very low prices) are more frequent than in a normal distribution.

**Low Kurtosis (<3)**: A kurtosis value significantly less than 3 would signify a platykurtic distribution, indicating lighter tails and a flatter peak. This implies that extreme values are less common than in a normal distribution.

**Kurtosis Around 3**: A kurtosis value close to 3 suggests a mesokurtic distribution, similar in shape to the normal distribution, indicating a moderate presence of extreme values.

Let's check the various methods for finding out outliers in the dataset.

* IQR Method

**IQR Method**

The IQR is the range between the first (Q1) and third (Q3) quartiles. Outliers are identified as values outside the range

```python
df
```

```python
# Specify numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Calculate IQR for each numeric column
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers based on IQR
df_outliers_iqr = df[((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

```

#### Handling outliers

Handling outliers in data involves identifying and addressing data points that significantly deviate from the overall pattern. This process aims to enhance the robustness and reliability of statistical analyses and modeling by reducing the influence of extreme values.

* Remove Outliers
* Transform Data
* Winsorizing
* Capping/Flooring

**Remove Outliers**

This method identifies and removes outliers based on the interquartile range (IQR), excluding data points beyond a certain range.
Useful when extreme values could significantly impact analysis, and their removal is considered appropriate.

```python
# Specify numeric columns
numeric_columns = df.select_dtypes(include='number').columns

# Calculate IQR for each numeric column
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers based on IQR
df_no_outliers = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(df_no_outliers.shape)
print(df_outliers_iqr.shape)
print(df.shape)

df_no_outliers.head(2)
```

#### Scaling and Normalization

Scaling and normalization are techniques used to standardize the range of independent variables or features of a dataset. Scaling ensures that all features contribute equally to the similarity or distance measures, preventing some features from dominating others.

Importance and Use:

Scaling and normalization are crucial in machine learning, especially for algorithms sensitive to the scale of input features. It improves convergence speed, avoids numerical instability, and ensures fair comparisons between different features.

There are 3 methods for scaling and normalization
* Min-Max Scaling
* Z-Score Normalization
* Robust Scaling

**Min-Max Scaling:**

Scales the data to a fixed range (usually [0, 1]).
Useful when the features have varying ranges.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

```python
# Min-Max Scaling

# Select only numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Apply Min-Max Scaling to numeric columns
min_max_scaler = MinMaxScaler()
df_min_max_scaled = df.copy()
df_min_max_scaled[numeric_columns] = min_max_scaler.fit_transform(df[numeric_columns])
```

**Standardization (Z-score normalization):**

Transforms data to have a mean of 0 and standard deviation of 1.
Suitable for algorithms expecting normally distributed features.

```python
# Apply Z-Score Normalization to numeric columns
z_score_scaler = StandardScaler()
df_z_score_normalized = df.copy()
df_z_score_normalized[numeric_columns] = z_score_scaler.fit_transform(df[numeric_columns])
```

Robust Scaling:

Scales features based on percentiles, making it robust to outliers.
Effective when dealing with datasets containing outliers

```python
## Apply Robust Scaling to numeric columns
robust_scaler = RobustScaler()
df_robust_scaled = df.copy()
df_robust_scaled[numeric_columns] = robust_scaler.fit_transform(df[numeric_columns])
```
