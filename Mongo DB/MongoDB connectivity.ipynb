```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient("localhost:27017") #Connect to MongoDB
```

```python
db = client['new_db'] # existing database
```

```python
collection = db['Insurance'] # existing collection in new_db
```

```python
data = collection.find()
type(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
# Display the bmi of age more than 40
data = collection.find({"age":{"$gt":40}},{"bmi":1})
df = pd.DataFrame(data)
df
```

```python
data = collection.find({"age":{"$gt":40}},{"bmi":1})
print(data)
for doc in data:
    print(doc)
```

```python
# Display the number of smokers per gender
data = collection.aggregate([{"$match":{"Smoker":"Yes"}},{"$group":{"_id":{"Gender":"$Gender"},"Count":{"$sum":1}}}])
df= pd.DataFrame(data)
df
```

```python
db.create_collection("coll")
```

```python
coll = db['coll']
coll.insert_many([{"Name":"Ann","Age":35,"Profession":"Teacher","Income":25000},
                { "Name":"Fatima","Age":42,"Profession":"Banker","Income":38000}])
```

```python
data = coll.find()
df = pd.DataFrame(data)
df
```

```python
# Change the income of Ann to 32000
coll.update_one({"Name":"Ann"},{"$set":{"Income":32000}})
```

```python
data = coll.find()
df = pd.DataFrame(data)
df
```

```python
coll.delete_one({"Name":"Ann"})
```

```python
data = coll.find()
df = pd.DataFrame(data)
df
```
